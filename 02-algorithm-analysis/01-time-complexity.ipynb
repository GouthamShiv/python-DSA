## `What is time complexity?`

- Amount of time taken by an algorithm to execute
- Runtime of an algorithm could be proportional
- There can be an increases in time based on the size of the input

## `Measuring time complexity`

There are two ways of analyzing time complexity
- Experimental Analysis: Runtime of the algorithm measured on various inputs
- Theoretical Analysis: Performed on description of the algorithm

## `Example 01` $\rightarrow$ sum of $n$ numbers

```c
int total = 0
int i = 1;
while (i <= n) {
  total = total + i;
  i = i + 1;
}
```

|Primitive Operations|Frequency|
|---|---|
|Declarations|$2$|
|Assignments|$2$|
|Comparison Operation|$N+1$|
|Arithmetic Operation|$N+N$|

$$2 + 2 + (N + 1) + (N + N) = 5 + 3N$$

## `Example 02` $\rightarrow$ computes the sum of a $m \times n$ matrix

```c
int total = 0;
for(int i=0; i<n; i++) {
  for(int j=0; j<m; j++) {
    total = total + A[i][j];
  }
}
```

|Primitive Operations|Frequency|
|---|---|
|Declarations|$1 + 1 + N$|
|Assignments|$1 + 1 + N$|
|Comparison Operation|$(N+1) + N \times (M+1)$|
|Accessing Array|$N \times M$|
|Increment Operation|$(N+1) + N \times (M+1)$|
|Arithmetic Operation|$N \times M$|

$$2(1+1+N) + 2((N+1) + N \times (M+1)) + 2(N \times M) \rightarrow 6 + 6N + 4NM = 6 + 6N + 4N^2$$
